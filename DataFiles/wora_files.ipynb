```python
# Import dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store CSV file in dataframe

wora_ages_csv = "WORA_ages_unpivoted.csv"
ages_df = pd.read_csv(wora_ages_csv)
ages_df.head()
```

```python
wora_ethn_csv = "WORA_ethnicity_unpivoted.csv"
ethn_df = pd.read_csv(wora_ethn_csv)
ethn_df.head()
```

```python
ages_group_df = ages_df.groupby(["state"]).count()
ages_group_df
```

```python
ethn_group_df = ethn_df.groupby(["state"]).count()
ethn_group_df
```

```python
new_ages_df = ages_df[['state', 'Age Group','Portion of WORA']].copy()
new_ages_df.head()
```

```python
n_ethn_df = ethn_df[['state', 'Race/Ethnicity','Portion of WORA']].copy()
n_ethn_df.head()
```

```python
new_ethn_df = n_ethn_df.rename(columns = {'Race/Ethnicity': 'Ethnicity'})
new_ethn_df
```

```python
# Connect to local database

protocol = 'postgresql'
username = 'postgres'
password = 'admin'
host = 'localhost'
port = 5432
database_name = 'wora_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
rds_connection_string
```

```python
# Checking Tables Names in PgAdmin

engine.table_names()
```

```python
new_ages_df.to_sql(name='wora_ages', con=engine, if_exists='replace', index=False)
```

```python
new_ethn_df.to_sql(name='wora_ethn', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from wora_ages', con=engine).head()
```

```python
pd.read_sql_query('select * from wora_ethn', con=engine).head()
```
